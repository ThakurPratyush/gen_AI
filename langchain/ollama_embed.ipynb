<h4> Ollama</h4>
Ollama supports embedding models making it possible to build retrival augmented generation RAG applications
that combine text prompts with exsisting documents or other data

```python
from langchain_community.embeddings import OllamaEmbeddings
```

```python
embeddings = (
    OllamaEmbeddings(model = "gemma2:2b") # by default uses llama2
    
)
```

```python
embeddings
```

```python
r1 = embeddings.embed_documents(
    [
        "alpha is the first letter of the greek alphabet",
        "beta is the second letter of greek alphabet"
    ]
)
```

```python
r1
```

```python
r1[0]
```

```python
embeddings.embed_query("what is the second letter of greek alphabet")
```
